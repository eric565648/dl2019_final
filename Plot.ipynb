```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

def show_plot(index,title,TRAIN_ACCU_1,TEST_ACCU_1,label_1,label_2,ylabel):
    plt.figure(figsize=(13,7))
    plt.title(title,fontsize=20)
    plt.xlabel('epoches',fontsize=18)
    plt.ylabel(ylabel,fontsize=18)
#     plt.xlim((index[0]-0.2, index[len(index)-1]+0.2))
#     plt.ylim((75, 85))
    plt.plot(index,TRAIN_ACCU_1,'r-',label=label_1)
    plt.plot(index,TEST_ACCU_1,'b-',label=label_2)
    #plt.plot(index,TRAIN_ACCU_2,'g-',label=label_3)
    #plt.plot(index,TEST_ACCU_2,'m-',label=label_4)
    #plt.plot(index,TRAIN_ACCURANCY_3,'k-',label='Train_Leaky_relu')
    #plt.plot(index,TEST_ACCURANCY_3,'m-',label='Test_Leaky_relu')
    plt.legend(loc='lower right',fontsize=12)
    plt.show()

# reload a file to a variable
with open('Train_loss.pickle', 'rb') as file:
    Train_loss =pickle.load(file)
#print(Train_loss)

with open('Train_Accu.pickle', 'rb') as file:
    Train_Accu =pickle.load(file)
#print(Train_Accu)

with open('Test_loss.pickle', 'rb') as file:
    Test_loss =pickle.load(file)
#print(Test_loss)

with open('Test_Accu.pickle', 'rb') as file:
    Test_Accu =pickle.load(file)
#print(Test_Accu)



index = np.arange(0,len(Train_loss),1)


show_plot(index,'',Train_loss,Test_loss,'Train loss','Test loss','loss')

show_plot(index,'',Train_Accu,Test_Accu,'Train Accurancy','Test Accurancy','accurancy(%)')
```
